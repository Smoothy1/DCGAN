```python
import pickle as pkl
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

import torch
from torchvision import datasets
from torchvision import transforms
```

```python
def get_dataloader(batch_size, image_size, data_dir=r"C:\Users\Миша\Downloads\Deep-Fake-Generator-master\content\processed_celeba_small\celeba"):
    """
    Batch the neural network data using DataLoader
    :param batch_size: The size of each batch; the number of images in a batch
    :param img_size: The square size of the image data (x, y)
    :param data_dir: Directory where image data is located
    :return: DataLoader with batched data
    """
    transform = transforms.Compose([transforms.Resize(image_size),transforms.CenterCrop(image_size),transforms.ToTensor()])
    # Реализуем функцию и возвращаем загрузчик данных
    dataset = datasets.ImageFolder(data_dir,transform = transform)
    
    dataloader = torch.utils.data.DataLoader(dataset = dataset,batch_size = batch_size,shuffle = True)
    return dataloader
```

```python
batch_size = 256
img_size = 32



celeba_train_loader = get_dataloader(batch_size, img_size)
```

***Предварительно обработаем данные изображения и масштабируем их в диапазоне пикселей от -1 до 1***


```python
def scale(x, feature_range=(-1, 1)):
    ''' Scale takes in an image x and returns that image, scaled
       with a feature_range of pixel values from -1 to 1. 
       This function assumes that the input x is already scaled from 0-1.'''
    # предположим, что x масштабируется до (0, 1)
    # масштабируем до feature_range и возвращаем масштаб x
    min, max = feature_range
    x = x*(max-min) + min
    return x
```

# ***Дискриминатор***


```python
import torch.nn as nn
import torch.nn.functional as F
```

```python
def conv(input_c,output,kernel_size,stride = 2,padding  = 1, batch_norm = True):
    layers =[]
    con = nn.Conv2d(input_c,output,kernel_size,stride,padding,bias = False)
    layers.append(con)
    
    if batch_norm:
        layers.append(nn.BatchNorm2d(output))
    
    return nn.Sequential(*layers)

class Discriminator(nn.Module):

    def __init__(self, conv_dim):
        """
        Initialize the Discriminator Module
        :param conv_dim: The depth of the first convolutional layer
        """
        super(Discriminator, self).__init__()
        self.conv_dim = conv_dim
        self.layer_1 = conv(3,conv_dim,4,batch_norm = False) #16
        self.layer_2 = conv(conv_dim,conv_dim*2,4) #8
        self.layer_3 = conv(conv_dim*2,conv_dim*4,4) #4
        self.fc = nn.Linear(conv_dim*4*4*4,1)
        # полная функция инициализации
        

    def forward(self, x):
        """
        Forward propagation of the neural network
        :param x: The input to the neural network     
        :return: Discriminator logits; the output of the neural network
        """
        # определяем поведение прямой связи
        x = F.leaky_relu(self.layer_1(x))
        x = F.leaky_relu(self.layer_2(x))
        x = F.leaky_relu(self.layer_3(x))
        x = x.view(-1,self.conv_dim*4*4*4)
        x = self.fc(x)
        return x
```

# ***Генератор***

Генератор должен выполнить выборку входных данных и сгенерировать новое изображение того же размера, что и наши обучающие данные 32x32x3. Это должно быть в основном транспонирование сверточных слоев с нормализацией, применяемой к выходным данным.

```python
def deconv(input_c,output,kernel_size,stride = 2, padding =1, batch_norm = True):
    layers = []
    decon = nn.ConvTranspose2d(input_c,output,kernel_size,stride,padding,bias = False)
    layers.append(decon)
    
    if batch_norm:
        layers.append(nn.BatchNorm2d(output))
    return nn.Sequential(*layers)

class Generator(nn.Module):
    
    def __init__(self, z_size, conv_dim):
        """
        Initialize the Generator Module
        :param z_size: The length of the input latent vector, z
        :param conv_dim: The depth of the inputs to the *last* transpose convolutional layer
        """
        super(Generator, self).__init__()
        self.conv_dim = conv_dim
        self.fc = nn.Linear(z_size,conv_dim*8*2*2)
        self.layer_1 = deconv(conv_dim*8,conv_dim*4,4)
        self.layer_2 = deconv(conv_dim*4,conv_dim*2,4)
        self.layer_3 = deconv(conv_dim*2,conv_dim,4)
        self.layer_4 = deconv(conv_dim,3,4,batch_norm = False)
        # полная функция инициализации
        
    def forward(self, x):
        """
        Forward propagation of the neural network
        :param x: The input to the neural network     
        :return: A 32x32x3 Tensor image as output
        """
        # определяем поведение прямой связи
        x = self.fc(x)
        x = x.view(-1,self.conv_dim*8,2,2) #(batch_size,depth,width,height)
        x = F.relu(self.layer_1(x))
        x = F.relu(self.layer_2(x))
        x = F.relu(self.layer_3(x))
        x = torch.tanh(self.layer_4(x))
        return x
```

**Все веса были инициализированы из нормального распределения с нулевым центром со стандартным отклонением 0,02.**

```python
def weights_init_normal(m):
    """
    Applies initial weights to certain layers in a model .
    The weights are taken from a normal distribution 
    with mean = 0, std dev = 0.02.
    :param m: A module or layer in a network    
    """
    # имя класса будет примерно таким:
    # `Conv`, `BatchNorm2d`, `Linear`, etc.
    classname = m.__class__.__name__
    
    if hasattr(m,'weight') and (classname.find('Conv') != -1 or classname.find('Linear') != -1):
        
        m.weight.data.normal_(0.0,0.02)
    
        if hasattr(m,'bias') and m.bias is not None:
            m.bias.data.zero_()
```

# ***Построение полной сети***

Определение гиперпараметров моделей и создание экземпляра дискриминатора и генератора из классов, определенных выше.

```python
def build_network(d_conv_dim, g_conv_dim, z_size):
    # определяем дискриминатор и генератор
    D = Discriminator(d_conv_dim)
    G = Generator(z_size=z_size, conv_dim=g_conv_dim)

    # инициализация весов модели
    D.apply(weights_init_normal)
    G.apply(weights_init_normal)

    print(D)
    print()
    print(G)
    
    return D, G
```

```python
d_conv_dim = 64
g_conv_dim = 64
z_size = 100

D, G = build_network(d_conv_dim, g_conv_dim, z_size)
```

```python
import torch

# Проверка GPU
train_on_gpu = torch.cuda.is_available()
if not train_on_gpu:
    print('No GPU found. Please use a GPU to train your neural network.')
else:
    print('Training on GPU!')
```

# ***Потери дискриминатора и генератора***

Теперь рассчитаем потери для обоих типов состязательных сетей.

**Потери дискриминатора**



*   Для дискриминатора общая потеря - это сумма потерь для реальных и поддельных изображений, d_loss = d_real_loss + d_fake_loss.



**Потери генератора**

Потеря генератора будет выглядеть аналогично только с перевернутыми метками. Цель генератора - заставить дискриминатор думать, что созданные им изображения реальны.

```python
def real_loss(D_out):
    '''Calculates how close discriminator outputs are to being real.
       param, D_out: discriminator logits
       return: real loss'''
    batch_size = D_out.size(0)
    labels = torch.ones(batch_size)
    if train_on_gpu:
        labels = labels.cuda()
    criterion = nn.BCEWithLogitsLoss()
    loss = criterion(D_out.squeeze(),labels)
    return loss

def fake_loss(D_out):
    '''Calculates how close discriminator outputs are to being fake.
       param, D_out: discriminator logits
       return: fake loss'''
    batch_size = D_out.size(0)
    labels = torch.zeros(batch_size)
    if train_on_gpu:
        labels = labels.cuda()
    criterion =  nn.BCEWithLogitsLoss()
    loss = criterion(D_out.squeeze(),labels)
    return loss
```

```python
import torch.optim as optim

# Создание оптимизаторов для дискриминатора D и генератора G
d_optimizer = optim.Adam(D.parameters(),lr = .0002, betas = [0.5,0.999])
g_optimizer = optim.Adam(G.parameters(),lr = .0002, betas = [0.5,0.999])
```

# ***Тренировка***

```python
def train(D, G, n_epochs, print_every=50):
    '''Trains adversarial networks for some number of epochs
       param, D: the discriminator network
       param, G: the generator network
       param, n_epochs: number of epochs to train for
       param, print_every: when to print and record the models' losses
       return: D and G losses'''
    
    # перемещение моделей на графический процессор
    if train_on_gpu:
        D.cuda()
        G.cuda()

    # будем отслеживать потери и сгенерированне "поддельные" образцы
    samples = []
    losses = []

    # Получаем некоторые фиксированные данные для выборки. Эти изображения
    # остаются неизменными на протяжении всего обучения и позволяют проверять производительность модели
    sample_size=16
    fixed_z = np.random.uniform(-1, 1, size=(sample_size, z_size))
    fixed_z = torch.from_numpy(fixed_z).float()
    # перемещение z на графический процессор, если он доступен
    if train_on_gpu:
        fixed_z = fixed_z.cuda()

    # цикл обучения периода
    for epoch in range(n_epochs):

        # цикл обучения пакета
        for batch_i, (real_images, _) in enumerate(celeba_train_loader):

            batch_size = real_images.size(0)
            real_images = scale(real_images)
            if train_on_gpu:
                real_images = real_images.cuda()
            # ===============================================
            #         CODE HERE: TRAIN THE NETWORKS
            # ===============================================
            
            # 1. Тренируйте дискриминатора на реальных и поддельных изображениях
            d_optimizer.zero_grad()
            d_out_real = D(real_images)
            z = np.random.uniform(-1,1,size = (batch_size,z_size))
            z = torch.from_numpy(z).float()
            if train_on_gpu:
                z = z.cuda()
            d_loss = real_loss(d_out_real) + fake_loss(D(G(z)))
            d_loss.backward()
            d_optimizer.step()
            # 2. Тренировка генератора с проигрышем в состязании
            G.train()
            g_optimizer.zero_grad()
            z = np.random.uniform(-1,1,size = (batch_size,z_size))
            z = torch.from_numpy(z).float()
            if train_on_gpu:
                z = z.cuda()
            g_loss = real_loss(D(G(z)))
            g_loss.backward()
            g_optimizer.step()
            
            # ===============================================
            #              END OF CODE
            # ===============================================

            # Выводим некоторые статистические данные о потерях
            if batch_i % print_every == 0:
                # добавляем потерю дискриминатора и потерю генератора
                losses.append((d_loss.item(), g_loss.item()))
                print('Epoch [{:5d}/{:5d}] | d_loss: {:6.4f} | g_loss: {:6.4f}'.format(
                        epoch+1, n_epochs, d_loss.item(), g_loss.item()))


        ## AFTER EACH EPOCH##    
        # создание и сохранение образцов, поддельных изображений
        G.eval() 
        samples_z = G(fixed_z)
        samples.append(samples_z)
        G.train()

    # Сохранение образцов обучающего генератора
    with open('train_samples.pkl', 'wb') as f:
        pkl.dump(samples, f)
    
    # возвращем потери
    return losses
```

```python
n_epochs = 30

losses = train(D, G, n_epochs=n_epochs)
```
